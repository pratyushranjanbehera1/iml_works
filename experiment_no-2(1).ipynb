```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python

x=np.array([[1,1,1],[1,1,2],[1,2,2]])
y=np.array([[6],[8],[9]])
x_transpose=np.transpose(x)
x_transpose
c = np.dot(x_transpose, x)
d=np.dot(x_transpose,y)
e=np.linalg.inv(c)
y_predicted=np.dot(e,d)
y_predicted
beta=y_predicted
print(beta)
b0 = beta[0][0]
b1 = beta[1][0]
b2 = beta[2][0]

print(f"y = {b0:.2f} + {b1:.2f}x1 + {b2:.2f}x2")
```
